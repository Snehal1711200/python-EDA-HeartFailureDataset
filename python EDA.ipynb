```python
import pandas as pd 
import seaborn as sns 
import matplotlib as plt

df= pd.read_csv('heart_failure_clinical_records_dataset (1).csv')
```

```python
#̄1.What is the distribution of age among heart failure patients in the dataset
 
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=20, kde=True, color='blue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()


#insights:
#Most of a people are from 50-70  age group range 
```

```python
# 2.How does the death rate vary with age
 
age_groups = pd.cut(df['age'], bins=[0, 40, 50, 60, 70, 80, 100])
death_rate_by_age = df.groupby(age_groups)['DEATH_EVENT'].mean() * 100

plt.figure(figsize=(8, 5))
death_rate_by_age.plot(kind='bar', color='red')
plt.title('Death Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Death Rate (%)')
plt.show()

#insights:
# death rate high in people who belong to 80 -100 age group range
```

```python

# 3.What is the percentage of male and female patients in the dataset?
 
gender_counts = df['sex'].value_counts(normalize=True) * 100  # Normalize=True gives percentages
labels = ['Male', 'Female'] if 1 in df['sex'].values else ['Female', 'Male']

plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'pink'])
plt.title('Gender Distribution')
plt.show()

#insights:
#percentage of males in dataset is 64.9 and female percentage is 35.1
```

```python
#4.How does the platelet count vary among different age groups?

age_bins = pd.cut(df['age'], bins=[0, 40, 50, 60, 70, 80, 100])
plt.figure(figsize=(8, 5))
sns.boxplot(x=age_bins, y='platelets', data=df)
plt.title('Platelet Count Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Platelet Count')
plt.show()

#insights:
#40 -50 age group having highest platelet count
```

```python
#5.Is there a correlation between creatinine and sodium levels in the blood<

correlation = df['serum_creatinine'].corr(df['serum_sodium'])
print(f"Correlation between serum creatinine and serum sodium: {correlation}")

plt.figure(figsize=(8, 5))
sns.scatterplot(x='serum_creatinine', y='serum_sodium', data=df, color='purple')
sns.regplot(x='serum_creatinine', y='serum_sodium', data=df, color='purple')

#insights:
#when sodium level is high creatinine level is low
```

```python
#6. How does the prevalence of high blood pressure differ between male and female patients<
 
hypertension_by_gender = df.groupby('sex')['high_blood_pressure'].mean() * 100
gender_labels = ['Female', 'Male']

plt.figure(figsize=(8, 5))
hypertension_by_gender.plot(kind='bar', color=['pink', 'blue'])
plt.title('Prevalence of High Blood Pressure by Gender')
plt.xticks([0, 1], gender_labels)
plt.ylabel('Percentage of Patients with High Blood Pressure')
plt.show()

#insights:
#female have highprevalence of high blood pressure
```

```python
#7. What is the relationship between smoking habits and the occurrence of heart failure<
 
smoking_impact = df.groupby('smoking')['DEATH_EVENT'].mean() * 100
smoking_labels = ['Non-Smoker', 'Smoker']

plt.figure(figsize=(8, 5))
smoking_impact.plot(kind='bar', color=['green', 'orange'])
plt.title('Death Event Percentage by Smoking Habit')
plt.xticks([0, 1], smoking_labels)
plt.ylabel('Percentage of Death Events')
plt.show()

#insights:
#non smoker having high rate of heart failure
```

```python
#8. Are there any noticeable patterns in the distribution of death events across different age groups<

age_bins = pd.cut(df['age'], bins=[0, 40, 50, 60, 70, 80, 100])
death_events_by_age = df.groupby(age_bins)['DEATH_EVENT'].mean() * 100

plt.figure(figsize=(8, 5))
death_events_by_age.plot(kind='bar', color='red')
plt.title('Distribution of Death Events by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Percentage of Death Events')
plt.show()

#insights:
# age the age increases death rate also increases
```

```python
#9. Is there any significant difference in ejection fraction between patients with and without diabetes<

plt.figure(figsize=(8, 5))
sns.boxplot(x='diabetes', y='ejection_fraction', data=df, palette=['blue', 'orange'])
plt.title('Ejection Fraction in Patients with and without Diabetes')
plt.xticks([0, 1], ['No Diabetes', 'Diabetes'])
plt.xlabel('Diabetes')
plt.ylabel('Ejection Fraction (%)')
plt.show()

#insights:
#not having with respect to value of mean
```

```python
#10. How does the serum creatinine level vary between patients who survived and those who did not?

sns.boxplot(x='DEATH_EVENT', y='serum_creatinine', data=df, palette=['green', 'red'])
plt.title('Serum Creatinine Levels: Survivors vs Non-Survivors')
plt.xticks([0, 1], ['Survived', 'Died'])
plt.xlabel('Death Event')
plt.ylabel('Serum Creatinine (mg/dL)')
plt.show()

#insights:
#people who were died having high creatinine level
```
